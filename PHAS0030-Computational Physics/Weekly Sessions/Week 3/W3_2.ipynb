```python
# In Class 2 Differentiation
#1)
def forward_difference(fun, x, dx):
    '''
    
    '''
    x = x + (dx/2)
    dfun = (fun(x + dx) - fun(x)) / dx
    return dfun

x_1 = 0.5
dx = 0.01
forward_difference(np.sin, x_1, dx)
```

```python
np.cos(x_1)
```

```python
dx = np.logspace(-5,1,10)
y_1 = []
for i in range(len(dx)):
    y_1.append(abs(np.cos(x_1)-forward_difference(np.sin, x_1, dx[i])))

fig = plt.figure(figsize=(9,3))
ax = fig.add_subplot(1,2,1)
ax.loglog(dx, y_1)
```

```python
#3) Second Derivative
def second_order_centred_difference(fun, x, dx):
    d2f = (fun(x+dx)-2*fun(x)+fun(x-dx))/(dx**2)
    return d2f
```

```python
dx = np.logspace(-6,-2,100)
y_1 = []
for i in range(len(dx)):
    y_1.append(abs((-np.sin(x_1))-second_order_centred_difference(np.sin, x_1, dx[i])))

fig = plt.figure(figsize=(9,3))
ax = fig.add_subplot(1,2,1)
ax.loglog(dx, y_1)
```

```python
def rectangle_int(fun,a,b,N):
    """
    f is a Python function to be integrated
    """
    dx = (b-a)/N
    
    I = []
    
    for i in range(N):
        term = fun(a + i*dx)*dx
        I.append(term)
    
    S = np.sum(I)
    return S
```

```python
def x1(x):
    return 2*x

rectangle_int(x1,0,1,10000)
```

```python
def x2(x):
    return x*np.cos(x)

rectangle_int(x2,0,1,10000)
```

```python
def trapezium_int(fun,a,b,N):
    """
    f is a Python function to be integrated
    """
    dx = (b-a)/N
    
    I = []
    
    for i in range(N):
        term = fun(a + i*dx)
        I.append(term)
    
    S = (np.sum(I) + (fun(b)/2) + (fun(a)/2))*dx
    return S
```

```python
trapezium_int(x2,0,1,10000)
```
