```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.logspace((10**(-3)),1,20)
y = np.zeros(20)
```

```python
print(x, y)
```

```python
a = 1
b = 1
theta = np.linspace(0,6*np.pi)
r = a + b*theta

fig = plt.figure(figsize=(9,3))
ax = fig.add_subplot(1,2,1)

ax.plot(x, y)
ax.grid(True)
ax.set_title('default grid')
```

```python
ax2 = fig.add_subplot(1,2,2)

ax2.semilogx(x, y)
ax.grid(True)
ax.set_title('default grid')
%matplotlib
```

```python
fig = plt.figure(figsize=(9,3))
ax = fig.add_subplot(1,2,1,polar=True)
ax.plot(theta, r)
```
