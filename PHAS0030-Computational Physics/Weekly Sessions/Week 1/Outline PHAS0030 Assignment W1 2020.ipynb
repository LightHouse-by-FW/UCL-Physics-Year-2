This notebook explores the electric potential and electric field casued by a monopole (1 charge), dipole (2 charges), and a quadrupole (4 charges) in 2 spatial dimensions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
spacing = 0.5
x = np.arange(0, 30, spacing)
y = np.arange(0, 30, spacing)

#particle 1 values
x_1 = 7
y_1 = 7
q_1 = 1

#particle 2 values
x_2 = 27
y_2 = 7
q_2 = -1

#particle 3 values
x_3 = 7
y_3 = 27
q_3 = -1

#particle 4 values
x_4 = 27
y_4 = 27
q_4 = 1
```

### Electrostatic Potential

The electrostatic potential energy from a singular charge at a point $ \textbf{r} $ is given by 

$$V(\textbf{r}) = k \frac{q}{\textbf{r} - \textbf{r}_0} $$

where the constant $$ k = \frac{1}{4\pi\epsilon_0}$$

and $ \textbf{r}_0 $ is the coordinates of the charged particle.

For the sake of efficiency in our calculations we will not be using he constant $ k $ since all scales will still be the same.

When looking at the electrostatic potential caused by two or more charges we simply calculate the potential for each charge seperately and then sum the values ofr each potential to get the resultant potential for that system.

The function below uses this equation to create a two dimensional array of floats to represent the electrostatic potential of a charged particle.

```python
def electricPotential2D(q_n, x_n, y_n, x, y):
    """
    creates a 2 dimensional array of the electric potential
    
    Inputs:
    q_n  Value of charge
    x_n, y_n  Location of charge (floats)
    x, y    Position(s) for potential to be evaluated
    """
    
    # adjusted x and y coordiante systems for the
    # location of the charge
    x_adj = (x - x_n)
    y_adj = (y - y_n)

    # 2 dimensional array of the distance from the charge
    x2D, y2D = np.meshgrid(x_adj, y_adj)
    r = np.sqrt(x2D*x2D + y2D*y2D)
    
    #order of the equation below
    order = 1
    
    # Calculate potential using q/r
    Ep = (q_n)/ r**order
    
    # return potential
    return Ep
```

### Electric Fields

The electric field created by a singular charge is

$$E(\textbf{r}) = kq \frac{(\textbf{r} - \textbf{r}_0)}{|\textbf{r} - \textbf{r}_0|^3} $$

where the constant $$ k = \frac{1}{4\pi\epsilon_0}$$.

and $ \textbf{r}_0 $ is the coordinates of the charged particle

When looking at the electric field caused by two or more charges we simply calculate the field components for each charge seperately and then sum the values for each component to get the resultant field component for that system.

The function below uses this equation to create 2 two dimensional arrays of floats to represent the x and y electric field components of a charged particle.

```python
def electricField2D(q_n, x_n, y_n, x, y):
    """
    creates 2 two dimensional arrays of floats to represent the
    x and y electric field components of a charged particle
    
    Inputs:
    q_n  Value of charge
    x_n, y_n  Location of charge (floats)
    x, y    Position(s) for potential to be evaluated
    """
    
    # adjusted x and y coordiante systems for the
    # location of the charge
    x_adj = (x - x_n)
    y_adj = (y - y_n)

    # 2 dimensional array of the distance from the charge
    x2D, y2D = np.meshgrid(x_adj, y_adj)
    r = np.sqrt(x2D*x2D + y2D*y2D)
    
    #order of the equation below
    order = 3
    
    # Calculate field x component using qx/r^3
    Ex = (q_n * x2D)/ r**order
    
    # Calculate field y component using qy/r^3
    Ey = (q_n * y2D)/ r**order
    
    # return Ex, Ey
    return Ex, Ey
```

```python
# Calculate x and y components of the electric field and
# potential for each charge
q1_Ex, q1_Ey = electricField2D(q_1, x_1, y_1, x, y)
q2_Ex, q2_Ey = electricField2D(q_2, x_2, y_2, x, y)
q3_Ex, q3_Ey = electricField2D(q_3, x_3, y_3, x, y)
q4_Ex, q4_Ey = electricField2D(q_4, x_4, y_4, x, y)
q1_Ep = electricPotential2D(q_1, x_1, y_1, x, y)
q2_Ep = electricPotential2D(q_2, x_2, y_2, x, y)
q3_Ep = electricPotential2D(q_3, x_3, y_3, x, y)
q4_Ep = electricPotential2D(q_4, x_4, y_4, x, y)
```

### Electrostatic Potential Plots

```python
# electric potential for one charge
E_total_p_1 = q1_Ep

# plot of corresponding electric potential
plt.imshow(E_total_p_1)
plt.xlabel("distance in x-direction")
plt.ylabel("distance in y-direction")
plt.title("Electric Potential for 1 charge")
plt.colorbar(label = "Electric Potential")
```

```python
# electric potential for two charges
E_total_p_2 = q1_Ep + q2_Ep

# plot of corresponding electric potential
plt.imshow(E_total_p_2)
plt.xlabel("distance in x-direction")
plt.ylabel("distance in y-direction")
plt.title("Electric Potential for 2 charges")
plt.colorbar(label = "Electric Potential")
```

```python
# electric potential for four charges
E_total_p_4 = q1_Ep + q2_Ep + q3_Ep + q4_Ep

# plot of corresponding electric potential
plt.imshow(E_total_p_4)
plt.xlabel("distance in x-direction")
plt.ylabel("distance in y-direction")
plt.title("Electric Potential for 4 charges")
plt.colorbar(label = "Electric Potential")
```

### Electric Field Plots

```python
# x and y components of the electric field for one charge
E_total_x_1 = q1_Ex
E_total_y_1 = q1_Ey

# streamplot and vector field of corresponding electric field
plt.quiver(x, y, E_total_x_1, E_total_y_1)
plt.streamplot(x, y, E_total_x_1, E_total_y_1)
plt.xlabel("distance in x-direction")
plt.ylabel("distance in y-direction")
plt.title("Electric Field for 1 charge")
```

```python
# x and y components of the electric field for two charges
E_total_x_2 = q1_Ex + q2_Ex
E_total_y_2 = q1_Ey + q2_Ey

# streamplot and vector field of corresponding electric field
plt.quiver(x, y, E_total_x_2, E_total_y_2)
plt.streamplot(x, y, E_total_x_2, E_total_y_2)
plt.xlabel("distance in x-direction")
plt.ylabel("distance in y-direction")
plt.title("Electric Field for 2 charges")
```

```python
# x and y components of the electric field for four charges
E_total_x_4 = q1_Ex + q2_Ex + q3_Ex + q4_Ex
E_total_y_4 = q1_Ey + q2_Ey + q3_Ey + q4_Ey

# streamplot and vector field of corresponding electric field
plt.quiver(x, y, E_total_x_4, E_total_y_4)
plt.streamplot(x, y, E_total_x_4, E_total_y_4)
plt.xlabel("distance in x-direction")
plt.ylabel("distance in y-direction")
plt.title("Electric Field for 4 charges")
```
