```python
# Seperating the function out into first order ODEs we get:
# dz/dt = v
# dv/dt = - g - (mu / m)*v

def first_order_derivative(input_array):
    """
    This function takes in a two entry array as input,storing 
    position and velocity values, and returns two first-order
    derivatives as an array.
    
    Input:
    input_array = two entry array
    
    Output:
    output_array = two first-order derivative array
    """
    
    output_array = np.copy(input_array)
    
    pos = input_array[0]
    vel = input_array[1]
    
    dz_dt = vel
    dv_dt = - g - mu*vel
    
    output_array[0] = dz_dt
    output_array[1] = dv_dt
    
    return output_array
```

```python
def sinh(x, N):
    """
    A python function to evaluate the expansion of sinh(x) to N terms.
    Where the expansion is...
    
    x^(2n + 1)/(2n + 1)!
    
    Input:
    x = point about which we are expanding
    N = Number of terms
    
    Output:
    sum_eval = value of the expansion
    """
    sum_eval = 0 # sum evaluation
    
    # for loop calculating each term and appending it to the list
    for n in range(N + 1):
    
        term = (x**(2*n + 1)/math.factorial(2*n + 1))
        
        sum_eval += term
    
    return sum_eval
```

```python
def forward_finite_difference(f, dx):
    """
    
    """
    fp1 = np.roll(f, 1)
    
    df_dx = (fp1 - f)/dx
    
    return df_dx
```

```python
dx = np.pi*(1/50)
x = np.linspace(0, np.pi*2 - dx)
sin_array = np.sin(x)
d_sin_array = forward_finite_difference(sin_array, dx)

plt.plot(x, sin_array)
plt.plot(x, d_sin_array)
```
